```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.drop(columns=['satisfaction']),df['satisfaction'],
                                                 test_size=0.2,random_state=42)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import StandardScaler
```

```python
#imputation transformer
trf1 = ColumnTransformer(transformers=[('Arrival Delay in Minutes',SimpleImputer(strategy='most_frequent'),[20])],remainder='passthrough')
```

```python
#onehotencoding
trf2 = ColumnTransformer(transformers=[('Types of Travel_class',OneHotEncoder(sparse_output = False,drop ='first', handle_unknown = 'ignore'),[3,4]),
                         ('Customer',OrdinalEncoder(categories= [['disloyal Customer','Loyal Customer']]),[1])]
                         ,remainder='passthrough')
```

```python
#trf3 = ColumnTransformer(transformers=[('num',StandardScaler(), [1,23])],remainder='passthrough')
```

```python
from sklearn.pipeline import Pipeline
pipe = Pipeline([('trf1',trf1),('trf2',trf2)])
```

```python
pipe.fit(x_train,x_test)
```

```python
x_train_transformed = pipe.fit_transform(x_train)
```

```python
x_test_transformed = pipe.fit_transform(x_test)
```

```python
scaler = StandardScaler()
scaler.fit(x_train_transformed)
```

```python
x_train_scaled = scaler.transform(x_train_transformed)
x_test_scaled = scaler.transform(x_test_transformed)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
y_train_transformed = label_encoder.fit_transform(y_train)
y_test_transformed = label_encoder.fit_transform(y_test)
```

```python
from sklearn.linear_model import LogisticRegression
logr= LogisticRegression()
logr.fit(x_train_scaled,y_train_transformed)
```

```python
logr.predict(x_test_scaled)
```

```python
r2score = logr.score(x_test_scaled,y_test_transformed)
print(r2score*100,"%")
```
